```python
def load_data(filename):
    path = Path("../data") / str("r" + filename + ".csv")
    data = pd.read_csv(path, header=None)
    data.columns = ["width", "height", "value"]
    return data
```

# Different dataset evaluation

```python
import pandas as pd
experiment_filepath = Path("experiments")

def load_summary_experiment(experiment_name):
    avg_scores = pd.read_csv(experiment_filepath / experiment_name / "avg_individual_scores.csv")
    best_scores = pd.read_csv(experiment_filepath / experiment_name / "best_individual_scores.csv")
    times = pd.read_csv(experiment_filepath / experiment_name / "times.csv")
    
    avg_scores.columns = ["index", "exp1", "exp2", "exp3"]
    best_scores.columns = ["index", "exp1", "exp2", "exp3"]
    times.columns = ["index", "exp1", "exp2", "exp3"]
        
    avg_scores.drop("index", axis=1, inplace=True)
    best_scores.drop("index", axis = 1, inplace = True)
    times.drop("index", axis=1, inplace = True)
    
    
    return best_scores, avg_scores, times
```

## 800
minimum score - 30000

```python
df = load_data("800")
```

```python
best_scores, avg_scores, times = load_summary_experiment("800_default")

plt.plot(best_scores.index, best_scores.mean(axis=1), label="best individual")
plt.plot(avg_scores.index, avg_scores.mean(axis=1), label="average in population")

plt.fill_between(best_scores.index, y1 = best_scores.mean(axis = 1) - best_scores.std(axis = 1), y2 = best_scores.mean(axis = 1) + best_scores.std(axis = 1), alpha = 0.2)
plt.fill_between(avg_scores.index, y1 = avg_scores.mean(axis = 1) - avg_scores.std(axis = 1), y2 = avg_scores.mean(axis = 1) + avg_scores.std(axis = 1), alpha = 0.2)


plt.axhline(y=30000, c="red", linestyle = "--", label = "minimum score")

plt.legend()
plt.title("Scores on dataset 800")
plt.savefig("../img/datasets/scores_800.png", format="png")
```

```python
print("{:.0f}".format(best_scores.mean(axis=1).iloc[-1]) +"±" + "{:.0f}".format(best_scores.std(axis=1).iloc[-1]),
      " & ", "{:.0f}".format(avg_scores.mean(axis=1).iloc[-1]) +"±" + "{:.0f}".format(avg_scores.std(axis=1).iloc[-1]))
```
